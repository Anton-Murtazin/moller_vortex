# Usage example

This notebook uses the `moller_vortex` package as a library. Definitions live in `src/`; this file is only for scans and inspection.


```python
import numpy as np
from moller_vortex import LGPacket, normalization_constant, S_impulse_closed_form
```

## Define packets and normalizations


```python
packet1 = LGPacket(ell=1, sigma_perp=0.18, sigma_par=0.35, kbar_z=20.0)
packet2 = LGPacket(ell=-1, sigma_perp=0.18, sigma_par=0.35, kbar_z=-20.0)

N1 = normalization_constant(packet1)
N2 = normalization_constant(packet2)
N1, N2
```

## Compute one impulse S-matrix element


```python
k3 = np.array([0.8, 0.10, 19.7])
k4 = np.array([-0.55, -0.08, -19.6])
impact_b = np.array([0.3, 0.0])

S, details = S_impulse_closed_form(
    k3,
    k4,
    packet1,
    packet2,
    lam1=0.5,
    lam2=0.5,
    lam3=0.5,
    lam4=0.5,
    impact_b=impact_b,
    N1=N1,
    N2=N2,
    return_details=True,
)
S, details['transverse_case']
```

## Scan over $\ell$


```python
scan = []
for ell in range(-3, 4):
    p1 = LGPacket(ell=ell, sigma_perp=0.18, sigma_par=0.35, kbar_z=20.0)
    N1_scan = normalization_constant(p1)
    S = S_impulse_closed_form(
        k3, k4, p1, packet2,
        lam1=0.5, lam2=0.5, lam3=0.5, lam4=0.5,
        impact_b=impact_b, N1=N1_scan, N2=N2,
    )
    scan.append((ell, S))

scan
```
